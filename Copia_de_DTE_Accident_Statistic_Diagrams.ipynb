## Simple visualizations

`.plot()` has several optional parameters. Most notably, the `kind` parameter accepts eleven different string values and determines which kind of plot to create:

* `"area"` is for area plots.
* `"bar"` is for vertical bar charts.
* `"barh"` is for horizontal bar charts.
* `"box"` is for box plots.
* `"hexbin"` is for hexbin plots.
* `"hist"` is for histograms.
* `"kde"` is for kernel density estimate charts.
* `"density"` is an alias for "kde".
* `"line"` is for line graphs.
* `"pie"` is for pie charts.
* `"scatter"` is for scatter plots.

See also: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.html

### Histogram

````python
df["USTUNDE"].plot(kind="hist") # Histogram based on the hour of the accidents
````

### Bar Chart

````python
# Count accidents by involved vehicle type
is_bicycle_total = df["IstRad"].sum()
is_car_total = df["IstPKW"].sum()
is_pedestrian_total = df["IstFuss"].sum()
is_bike_total = df["IstKrad"].sum()
is_truck_total = df["IstGkfz"].sum()
is_other_total = df["IstSonstige"].sum()

# create a new DataFrame
accident_types = [is_bicycle_total, is_car_total, is_pedestrian_total, is_bike_total, is_truck_total, is_other_total]
index = ['Bicycle', 'Car', 'Pedestrian', 'Bike', 'Truck', 'Other']
df_accident_types = pd.DataFrame({'Number of Accidents': accident_types}, index=index)

# create bar chart
df_accident_types = df_accident_types.sort_values(by=['Number of Accidents'], ascending=False)
df_accident_types.plot(kind='bar')
````

### Pie Chart

````python
df_accident_types.plot(kind='pie', y='Number of Accidents', figsize=(10,10))
````

### Line Chart

````python
# grouping rows with the same month and counting them
df_month = df.groupby(by='UMONAT').count()

df_month['OBJECTID'].plot(kind='line',                 # Line chart
                          xticks=df_month.index,       # Displays every value on X
                          xlabel='Month',              # Label for x-Axis
                          ylabel='Number of Accidents',# Label for y-Axis
                          title='Accidents per Month', # Headline of the diagram
                          figsize=(10,10))             # Size of the diagram
````

### Bar Chart (horizontal)

````python
# In order to show proper labels for each month, we need a mapping
month_mapping = {1: 'Jan',
                 2: 'Feb',
                 3: 'Mar',
                 4: 'Apr',
                 5: 'May',
                 6: 'Jun',
                 7: 'Jul',
                 8: 'Aug',
                 9: 'Sept',
                 10: 'Oct',
                 11: 'Nov',
                 12: 'Dec', }

# Adding a new column in DataFrame with month name
df_month['Month Name'] = df_month.index.map(month_mapping)

df_month['OBJECTID'].plot(kind='barh', figsize=(10,10)).set_yticklabels(df_month['Month Name'])
# Note: set_yticklabels sets changes the axis labels to the month name
````

# New Tpye of Diagrams


# **Blox Plot**

````python
plt.figure(figsize=(10, 5))
plt.boxplot(df['USTUNDE'], vert=False)
plt.title("Box Plot of USTUNDE")
plt.show()
````

# **Area Plot**

````python
# Grouping the data by UMONAT and calculating the mean of USTUNDE
ustunde_by_umonat = df.groupby('UMONAT')['USTUNDE'].mean()

# Plotting as a bar chart
ustunde_by_umonat.plot(kind='bar', color='skyblue')
plt.title("Average USTUNDE by UMONAT")
plt.xlabel("UMONAT")
plt.ylabel("Average USTUNDE")
plt.grid(axis='y')  # Add grid lines on the y-axis
plt.show()
````

# Violin Plot

````python
plt.figure(figsize=(10, 5))
sns.violinplot(x=df['ULAND'], y=df['USTUNDE'])
plt.title("Violin Plot of USTUNDE by ULAND")
plt.show()
````

## **Heatmap**

````python
pivot_data = df.groupby(["UMONAT", "ULAND"])["USTUNDE"].mean().unstack()
plt.figure(figsize=(10, 5))
pivot_data = pivot_data.astype(int)
sns.heatmap(pivot_data, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Heatmap of USTUNDE by UMONAT and ULAND")
plt.show()
````
